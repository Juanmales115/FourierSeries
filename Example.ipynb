```python
from src.FourierFunct import animar_fourier, f_periodic
import numpy as np
```

```python
def f_base(x):
    return np.where(x <= 0, np.abs(x), np.sqrt(x))
def f(x):
    return f_periodic(f_base, x)

animar_fourier(f, N_MAX=30, A=-2*np.pi, B=2*np.pi, filename="fourierex1.gif", fps=5)
```

```python
def f_base(x):
    return np.where(x < 0, 1/x, np.log(x+0.001))

def f(x):
    return f_periodic(f_base, x)

animar_fourier(f, N_MAX=200, A=-2*np.pi, B=2*np.pi, filename="fourierex2.gif", fps=15, h_min=0.1, h_max=5)
```

```python
def f_base(x):
    return np.where(x < 0, np.sqrt(-x), np.sqrt(x))

def f(x):
    return f_periodic(f_base, x)

animar_fourier(f, N_MAX=50, A=-2*np.pi, B=2*np.pi, filename="fourierex3.gif", fps=8)
```
