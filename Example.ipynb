```python
import src.FourierFunct as ff
import numpy as np
```

```python
def f_base(x):
    return np.where(x <= 0, np.abs(x), np.sqrt(x))
def f(x):
    return ff.f_periodic(f_base, x)

ff.animate_fourier(f, N_MAX=30, A=-2*np.pi, B=2*np.pi, filename="fourierex1.gif", fps=5)
```

```python
ff.animate_frejer(f, N_MAX=30, A=-2*np.pi, B=2*np.pi, filename="frejerex1.gif", fps=5)
```

```python
def f_base(x):
    return np.where(x < 0, 1/x, np.log(x+0.001))

def f(x):
    return ff.f_periodic(f_base, x)

ff.animate_fourier(f, N_MAX=200, A=-2*np.pi, B=2*np.pi, filename="fourierex2.gif", fps=15, h_min=0.1, h_max=5)
```

```python
ff.animate_frejer(f, N_MAX=200, A=-2*np.pi, B=2*np.pi, filename="frejerex2.gif", fps=15, h_min=0.1, h_max=5)
```

```python
def f_base(x):
    return np.where(x < 0, np.sin(x), np.cos(x))

def f(x):
    return ff.f_periodic(f_base, x)

ff.animate_fourier(f, N_MAX=200, A=-2*np.pi, B=2*np.pi, filename="fourierex3.gif", fps=30, h_min=1.5, h_max=2)
```

```python
ff.animate_frejer(f, N_MAX=200, A=-2*np.pi, B=2*np.pi, filename="frejerex3.gif", fps=30, h_min=1.5, h_max=2)
```
